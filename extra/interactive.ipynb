```julia
xs = range(0,stop=25,length=100)
@manipulate for k in range(0.5,stop=10,length=51), theta in range(0.1,stop=3,length=51)
    f = Distributions.Gamma(k,theta)
    ys = [Distributions.pdf(f,x) for x in xs]
    p = plot(xs,ys)
    ylims!(0,0.2)
    p
end
```

```julia
function model_p_i1(x::Number)::Float64
    f = Distributions.Gamma(2.84,5.05)
    return Distributions.pdf(f,x)
end
n = 1000
prob_d = 0.1
prob_i = 0.055
c = 50
l = Int(ceil(n*f))
suc = [Agent(suceptible,0,x->prob_i,model_p_i1,x->prob_d) for i in 1:l]
inf = [Agent(infected,0,x->prob_i,model_p_i1,x->prob_d) for i in l+1:n]
N = System(n,vcat(suc,inf),watts_strogatz(n,c,0.5))
res = basicSimulation(N,500,advanceParallel!,(i,N)->evolve(i,N,i_tmin=15,r_tmin=90))
res = [Int(r) for r in res]
p = plot()
plotEvolution(res)
p
```
